```
df = spark.read.csv('/mnt/shirleycho-group3/Hotel_data/new_york_hotels.csv',header=True,inferSchema=True)
display(df)
pdf = df.toPandas()
display(pdf)

def getnames2(x):
    mydict = {'1':'Manhattan',
            '2':'Bronx',
            '5':'Staten island',
            '4':'Queens',
            '3':'Brooklyn'}
    try:
        if int(x) >= 10001 and int(x) <=10282:
            return mydict.get('1')
        elif int(x) >= 10301 and int(x) <=10314:
            return mydict.get('5')
        elif int(x) >= 10451 and int(x) <=10475:
            return mydict.get('2')
        elif int(x) >= 11004 and int(x) <=11109:
            return mydict.get('4')
        elif int(x) >= 11351 and int(x) <=11697:
            return mydict.get('4')
        elif int(x) >= 11201 and int(x) <=11256:
            return mydict.get('3')
        else:
            return None
    except:
        return None
def getints(x):
    try:
        val = int(x)
    except:
        val = 0
    return val
```

```
pdf['postalcodeint'] = pdf['postal_code'].map(lambda x: getints(x))
```

```
pdf['Neighborhood']=pdf['postalcodeint'].map(lambda x: getnames2(x))

pdf.drop(columns=['ean_hotel_id'],inplace=True)

pdf = pdf.dropna(subset = ['Neighborhood'])

display(pdf)
print(pdf.info())
```

```
def turntofloat(x):
    try:
        val = float(x)
    except:
        val = 0
    return val
pdf['star_rating'] = pdf['star_rating'].map(lambda x: turntofloat(x))

display(pdf)
print(pdf.info())
```

```
pdf.drop(columns=['postalcodeint'],inplace=True)
pdf.rename({"address1":"address"},axis=1,inplace=True)
cols = pdf.columns
for c in cols:
    pdf.rename({c:"hotel_"+c},axis=1,inplace=True)

display(pdf)
print(pdf.info())
```

```
sparkDF=spark.createDataFrame(pdf)

display(sparkDF)
sparkDF.printSchema()
sparkDF.write.mode("overwrite").options(header='True').csv('/mnt/shirleycho-group3/Hotels_clean/Hotels')
```
