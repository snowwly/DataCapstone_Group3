```
df = spark.read.csv('/mnt/shirleycho-group3/Zillow/Metro_new_listings_uc_sfrcondo_sm_month.csv',header=True,inferSchema=True)
display(df)
```

```
df.printSchema()
```

```
pdf = df.toPandas()
display(pdf)
```

```

pdf = pdf[pdf['RegionName'].map(lambda x:x=='New York, NY')]

display(pdf)

```

```
pdf = pdf.drop(columns = ['RegionID','SizeRank','RegionType','StateName'])
display(pdf)
```

```
row1= pdf[0:1]
cols = row1.columns[2:-1].tolist()
#print(cols)
row1_unpivot = pd.melt(row1,id_vars='RegionName',value_vars=cols)
display(row1_unpivot)
row1_unpivot.rename({'variable':'Date'},axis=1,inplace=True)
row1_unpivot['Year'] = row1_unpivot['Date'].map(lambda x:x.split('-')[0])
row1_unpivot['Month']=row1_unpivot['Date'].map(lambda x:x.split('-')[1])
row1_unpivot.rename({'value':'Count_of_Homes_Listed'},axis=1,inplace=True)

display(row1_unpivot)
newdf=row1_unpivot



```

```
newdf.info()
newdf['Year'] = newdf['Year'].astype(int)
newdf['Month'] = newdf['Month'].astype(int)
newdf['Count_of_Homes_Listed'] = newdf['Count_of_Homes_Listed'].astype(int)
newdf.info()
display(newdf)
```
