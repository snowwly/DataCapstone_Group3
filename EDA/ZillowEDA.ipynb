```python
# display(dbutils.fs.ls("/mnt/shirleycho-group3/Zillow"))
import pandas as pd
```

```python
pdf = pd.read_csv(r'./Data/County_zori_sm_month.csv')


display(pdf)
print(pdf.shape)
```

```python
display(pdf[pdf['State']=='NY']) #weird cause there hsould be 62 new york counties but we only need the 5 for new york city
```

```python
# Manhattan = New York County.
# Brooklyn = Kings County.
# Queens = Queens County.
# Staten Island = Richmond County.
# The Bronx = Bronx County.
def getonlyNYC(x):
    x = x.lower()
    if x == 'bronx county' or x == 'kings county' or x == 'new york county' or x =='queens county' or x == 'richmond county':
        return True
    return False
pdf = pdf[pdf['RegionName'].map(lambda x: getonlyNYC(x))]

display(pdf) #theres another county in GA thats the same name. get rid of that.

pdf= pdf[pdf['State']=='NY']
display(pdf)
```

```python
row1 = pdf[0:1]
display(row1)
cols = row1.columns[9:-1].tolist()
print(cols)
row1_unpivot = pd.melt(row1,id_vars='RegionName',value_vars=cols)
print(row1_unpivot)
row1_unpivot.info()
print(row1_unpivot['value'].max())
print(row1_unpivot['value'].min())
```

```python
row2 = pdf[1:2]
display(row2)
cols = row2.columns[9:-1].tolist()
print(cols)
row2_unpivot = pd.melt(row2,id_vars='RegionName',value_vars=cols)
print(row2_unpivot)
row2_unpivot.info()
print(row2_unpivot['value'].max())
print(row2_unpivot['value'].min())
```

```python
row3 = pdf[2:3]
display(row3)
cols = row3.columns[9:-1].tolist()
print(cols)
row3_unpivot = pd.melt(row3,id_vars='RegionName',value_vars=cols)
print(row3_unpivot)
row3_unpivot.info()
print(row3_unpivot['value'].max())
print(row3_unpivot['value'].min())
```

```python
row4 = pdf[3:4]
display(row4)
cols = row4.columns[9:-1].tolist()
print(cols)
row4_unpivot = pd.melt(row4,id_vars='RegionName',value_vars=cols)
print(row4_unpivot)
row4_unpivot.info()
print(row4_unpivot['value'].max())
print(row4_unpivot['value'].min())
```

```python
row5 = pdf[4:5]
display(row5)
cols = row5.columns[9:-1].tolist()
print(cols)
row5_unpivot = pd.melt(row5,id_vars='RegionName',value_vars=cols)
print(row5_unpivot)
row5_unpivot.info()
print(row5_unpivot['value'].max())
print(row5_unpivot['value'].min())
```

```python
newdf = pd.concat([row1_unpivot,row2_unpivot,row3_unpivot,row4_unpivot,row5_unpivot],axis=0)
print(newdf['value'].max())
print(newdf['value'].min())
```

```python
import plotly.express as px

fig = px.line(newdf, x="variable", y="value",  title='avg rental house prices over time',color='RegionName')
fig.show('png')
```
