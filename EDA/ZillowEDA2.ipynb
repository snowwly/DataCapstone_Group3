```python

import pandas as pd
```

```python
pdf = pd.read_csv(r'./Data/Metro_new_listings_uc_sfrcondo_sm_month.csv')
print(pdf.info())
display(pdf)
print(pdf.shape)
```

```python
pdf = pdf[pdf['RegionName'].map(lambda x:x=='New York, NY')]

display(pdf)
pdf.info()

```

```python
row1= pdf[0:1]
cols = row1.columns[5:-1].tolist()
print(cols)
row1_unpivot = pd.melt(row1,id_vars='RegionName',value_vars=cols)
display(row1_unpivot)

print(row1_unpivot['value'].max())

print(row1_unpivot['value'].min())
```

```python
import plotly.express as px
fig = px.line(row1_unpivot, x="variable", y="value", title='zillow housing listing count per month from 2018 to 2023')
fig.show("png")
```

```python
row1_unpivot['year'] = row1_unpivot['variable'].map(lambda x: x.split("-")[0])


newdf = row1_unpivot.groupby('year').agg('sum')['value']
display(newdf)


fig = px.bar(newdf, x=newdf.index, y='value')
fig.show("png")

#might have to ignore 2023, because its incomplete in a bar graph
fig = px.bar(row1_unpivot, x='variable', y='value')
fig.show("png")

```
